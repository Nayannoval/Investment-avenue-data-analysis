```python
import pandas as pd
import numpy as np
```

```python
import os
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('Finance_data.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.drop(['Mutual_Funds', 'Equity_Market' , 'Debentures','Government_Bonds','PPF','Fixed_Deposits','Gold'], axis=1, inplace = True)
```

```python
df.replace(r'^\s*$',np.nan, regex= True)
```

```python
sns.countplot( data = df, x='gender' , linewidth =1, palette ='Set2',edgecolor='black')
plt.show()
```

```python
sns.countplot(data =df , x='age',linewidth =3, palette='Set3',edgecolor ='black')
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.countplot(x=df['gender'],hue=df['Investment_Avenues'],palette='summer' , linewidth=3, edgecolor='white' )
plt.title('INVESTMENT AVENUE')

plt.subplot(1,2,2)
sns.countplot(x=df['gender'],hue=df['Stock_Marktet'],palette='hot' , linewidth =3 , edgecolor='white')
plt.title('STOCK MARKET')

plt.show()
```

```python
sns.countplot(x=df['Factor'], palette='coolwarm' , linewidth=3,edgecolor='black')
plt.show()
```

```python
plt.figure(figsize=(18,6))
sns.pointplot( x='Purpose' , y='age' , data =df,capsize=0.5,color ='Blue',linestyles='--')
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.countplot(data =df , hue = df['gender'], x=df['Duration'],linewidth=2,edgecolor='black')
plt.title('DURATION')

plt.subplot(1,2,2)
sns.countplot(hue=df['gender'],x=df['Invest_Monitor'],palette ='seismic',linewidth=2,edgecolor='black')
plt.title('INVESTMENT MONITORING')
```
